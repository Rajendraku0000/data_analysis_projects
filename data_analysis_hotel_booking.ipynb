 ##  data analysis and visualization

```python
cencel_pernc=df["is_canceled"].value_counts(normalize=True)
print(cencel_pernc)

plt.figure(figsize=(5,4))
plt.bar(["not cancel","cencel"],df["is_canceled"].value_counts(normalize=True))
plt.title("reservation status count")
plt.show()
```

```python
df.head()
```

```python
df["hotel"].value_counts(normalize=True)
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x="hotel",hue="is_canceled",data=df)
plt.title("reservation status in different hotel")
plt.show()
```

```python
resort_hotel=df[df["hotel"]=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
city_hotel=df[df["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize=(10,5))
plt.title("average daily range in city and resort hotel")
plt.plot(resort_hotel.index,resort_hotel["adr"],label="resort_hotel")
plt.plot(city_hotel.index,city_hotel["adr"],label="city_hotel")
plt.show()
```

```python
df["month"]=df["reservation_status_date"].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x="month",hue="is_canceled",data=df)
plt.title("reservation status per month")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("adr per month")
sns.barplot("month","adr",data=df[df["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index())
plt.show()
```

```python
cancelled_data=df[df["is_canceled"]==1]
top_10_country=cancelled_data["country"].value_counts()[:10]
```

```python
# plt.figure(figsize=(8,8))
plt.pie(top_10_country,autopct="%.2f",labels=top_10_country.index)
plt.title("top 10 coutry with cancelled data")
```
